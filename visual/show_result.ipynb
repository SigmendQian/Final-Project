```python
import matplotlib.pyplot as plt
import json
import os
import numpy as np
```

```python
seeds = [100,200,300]
lenets = [json.load(open(f'../results/lenet_{seed}_record.json')) for seed in seeds]
vggs = [json.load(open(f'../results/vgg_{seed}_record.json')) for seed in seeds]
```

```python
# 绘制train loss曲线

## 读取数据，求均值和方差
lenet_train_loss = np.array([info['train_loss'] for info in lenets])
lenet_train_loss_mean = np.mean(lenet_train_loss,axis=0).reshape(-1) # 求均值
lenet_train_loss_std = np.std(lenet_train_loss,axis=0).reshape(-1) # 求方差
vgg_train_loss = np.array([info['train_loss'] for info in vggs]) 
vgg_train_loss_mean = np.mean(vgg_train_loss,axis=0).reshape(-1) # 求均值
vgg_train_loss_std = np.std(vgg_train_loss,axis=0).reshape(-1) # 求均值

## 画图
x = range(len(vgg_train_loss_mean)) # 横坐标
plt.plot(x, lenet_train_loss_mean, color='red',label='LeNet')
# fill_between是绘制阴影区间
plt.fill_between(x, lenet_train_loss_mean - lenet_train_loss_std, lenet_train_loss_mean+lenet_train_loss_std, color='red', alpha=0.5) 
# fill_between是绘制阴影区间
plt.plot(x, vgg_train_loss_mean, color='black',label='Vgg')
plt.fill_between(x, vgg_train_loss_mean - vgg_train_loss_std,vgg_train_loss_mean + vgg_train_loss_std, color='black', alpha=0.5)

# 添加标题和
plt.xlabel('Epoch')
plt.ylabel('Train Loss')
plt.legend()

# 显示图形
plt.grid(True)
plt.show()
```

```python
# 绘制test loss曲线

## 读取数据，求均值和方差
lenet_test_loss = np.array([info['test_loss'] for info in lenets])
lenet_test_loss_mean = np.mean(lenet_test_loss,axis=0).reshape(-1) # 求均值
lenet_test_loss_std = np.std(lenet_test_loss,axis=0).reshape(-1) # 求方差
vgg_test_loss = np.array([info['test_loss'] for info in vggs]) 
vgg_test_loss_mean = np.mean(vgg_test_loss,axis=0).reshape(-1) # 求均值
vgg_test_loss_std = np.std(vgg_test_loss,axis=0).reshape(-1) # 求均值

## 画图
x = range(len(vgg_test_loss_mean)) # 横坐标
plt.plot(x, lenet_test_loss_mean, color='red',label='LeNet')
plt.fill_between(x, lenet_test_loss_mean - lenet_test_loss_std, lenet_test_loss_mean+lenet_test_loss_std, color='red', alpha=0.5)
plt.plot(x, vgg_test_loss_mean, color='black',label='Vgg')
plt.fill_between(x, vgg_test_loss_mean - vgg_test_loss_std,vgg_test_loss_mean + vgg_test_loss_std, color='black', alpha=0.5)

## 添加标题和
plt.xlabel('Epoch')
plt.ylabel('Train Loss')
plt.legend()

## 显示图形
plt.grid(True)
plt.show()
```

```python
# 绘制test loss曲线

## 读取数据，求均值和方差
lenet_acc = np.array([info['accuracy'] for info in lenets])
lenet_mean = np.mean(lenet_acc,axis=0).reshape(-1) # 求均值
lenet_std = np.std(lenet_acc,axis=0).reshape(-1) # 求方差
vgg_acc = np.array([info['accuracy'] for info in vggs]) 
vgg_mean = np.mean(vgg_acc,axis=0).reshape(-1) # 求均值
vgg_std = np.std(vgg_acc,axis=0).reshape(-1) # 求均值
## 画图
x = range(len(vgg_mean)) # 横坐标
plt.plot(x, lenet_mean, color='red',label='LeNet')
plt.fill_between(x, lenet_mean - lenet_std, lenet_mean+lenet_std, color='red', alpha=0.5)
plt.plot(x, vgg_mean, color='black',label='Vgg')
plt.fill_between(x, vgg_mean - vgg_std,vgg_mean + vgg_std, color='black', alpha=0.5)

## 添加标题和
plt.xlabel('Epoch')
plt.ylabel('Train Loss')
plt.legend()

## 显示图形
plt.grid(True)
plt.show()
```
